```python
import pandas as pd
from pprint import pprint
import numpy as np
import tensorflow as tf

import matplotlib.pyplot as plt

df = pd.read_csv('ozone.csv')
pprint(df)

#df = df.replace(np.nan, 0)

# 아웃라이어 결측치 제거시
# 제거되는걸 방지하기 위해 필요한 column만 추출
df = df[["Ozone", "Temp"]]

# 결측치를 제거
df = df.dropna(how="any", inplace=False)

pprint(df)
```

```python
# linear regression
# 1. training data set 준비
# # 데이터 사이즈 커질 수록 

x_data = df["Temp"] # [1.2.3]
y_data = df["Ozone"] 

# 2. 입력을 받아들이는 노드의 플레이스 호더
x = tf.placeholder(dtype=tf.float32)
y = tf.placeholder(dtype=tf.float32)

# 3. 웨이트 바이어스 정의
W = tf.Variable(tf.random_normal([1]), name="weight")
b = tf.Variable(tf.random_normal([1]), name="bias")

# 4. 가설 하이포세시스
H = W * x + b

# 5. cost func 정의
cost = tf.reduce_mean(tf.square(H-y))

# 6. train
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)

# 7. session

sess = tf.Session()
sess.run(tf.global_variables_initializer())

# 8. 학습

for step in range(3000):
    _,cost_val = sess.run([train,cost], feed_dict={ x : x_data, 
                                y : y_data})

    if step % 300 == 0:
        print( "cost_val : {}".format( cost_val))
        
## 정제된 데이터를 정규화나 표준화를 이용해서 값의 범위
# 조절해야함 그래야 학습이 정상적으로 이루어짐
# 데이터를 표준화 방식으로 다시 만들어서 학습을 진행
# 노말리제이션 (표준화):
# 각 데이터를 다음으로 같은 방식으로 비율적으로 축소시키는 방식
# 각 요소 값 -> (요소값 - 전체데이터의 최소값) / (전체 데이터의 최대값 - 전체 데이터의 최소값)

df["Temp_Norm"] = (df["Temp"] - df["Temp"].min())/ (df["Temp"].max() - df["Temp"].min())
df["Ozone_Norm"] = (df["Ozone"] - df["Ozone"].min())/ (df["Ozone"].max() - df["Ozone"].min())

# 확인작업
plt.scatter(df["Temp_Norm"],df["Ozone_Norm"])

print(df["Temp_Norm"])
```
